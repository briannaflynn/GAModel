```python
import numpy as np
import pandas as pd
from gamodel import GAModel

np.random.seed(0)

# Number of samples
n_samples = 100

# Create a linearly spaced X variable
X1 = np.linspace(-10, 10, n_samples)

# Create a second X variable that is a random normal distribution
X2 = np.random.normal(size=n_samples)

# Create a third X variable that is a random uniform distribution
X3 = np.random.uniform(-1, 1, size=n_samples)

# Create y such that it has an approximately linear relationship with X1, X2, and X3
y = 3*X1 - 1.5*X2 + 2*X3 + np.random.normal(scale=3, size=n_samples)

# Create a DataFrame
df = pd.DataFrame({'X1': X1, 'X2': X2, 'X3': X3, 'y': y})

df
```

```python
target = df['y'].to_numpy()
```

```python
target
```

```python
# Create and fit a GAModel
model = GAModel(df[['X1', 'X2', 'X3']].values, df['y'].values, spline_index=0, n_splines=20, linear_indices=[1, 2])
model.fit_model()

# Plot the results
model.plot_results()
```
